## Changing the List

A cool thing with a list is that you can add, delete or modify existing values.

### append
A useful way to add a new entry lists is applying the append function. 



```python
x = [1,1.5,"hello world",True]
x.append(0) # adds a single item
print(x)
```

### extend
Extend function is used for joining two lists.


```python
# Create a list (call it y) with entries 1.5 and '2020'
y = [1.5,'2020'] 
```

```python
# Join x and y 
x.extend(y) 
print(x)
#x = [1, 1.5, 'hello world', True, 0, 1.5, '2020']
```

### remove
You can remove all items by clear( ). If you want to remove specific values in the list, apply remove( ). 
To remove an item by index, use pop() or the del statement.

```python
# x.clear() removes all of the items
x.remove('2020') # removes the '2020'
x.remove(1.5) # removes the first value of 1.5
x.remove(1.5) # removes the second value of 1.5
print(x)
```

```python
# Other way to remove things
del x[0]  # removes first entry, 1
print(x)
```

```python
x.pop(2)  # removes third entry, 0
print(x)
```

```python
x.pop() # removes last entry, True
print(x)
```

`range(b) is a shorthand where a = 0 and c = 1. 

```python
# This tells you nothing and doesn't show what's in the sequence
range(10) 
 
```

```python
# Range can be confusing because shows the starts with zero.  
# to print
x = [i for i in range(10)] # 0 to 3 
x
```

```python
# More advanced code: 
# This evaluates if the array of numbers is <5, =5 or >5 
x = [0,5,10]
for i in range(len(x)):
    if x[i] == 5: # we use two equals for comparison, since one equals tells python to set x to that value
        print("x = 5")
    elif x[i] > 5: # this is an else-if statement. It allows us to do multiple checks in one if setup
        print("x > 5")
    else:
        print("x < 5")
```

## Let's plot 

```python
# plot a cubic function between -e and e (2.718281828459045)
import math
import matplotlib.pyplot as plt # this is a standard convention when using matplotlib, we name it plt.
import numpy as np  

print(math.e) # 2.718281828459045

# Range only takes integer values so linspace is useful for dividing an interval into different points
x = np.linspace(-math.e,math.e,50) # takes 50 values from the interval [-1,1]
y = x**3
plt.plot(x,y); # this reads the two lists and grabs element 0 from x and y, then draws it.  
plt.xlabel("X"); 
plt.ylabel("Y = X^3");
plt.title("Cubic Function");
```
